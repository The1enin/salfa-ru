# Cоздать прогностическую модель рисков беременных
## Описание данных

Age: Возраст в годах

SystolicBP: Верхнее значение артериального давления в мм рт. ст.

DiastolicBP: Более низкое значение артериального давления в мм рт. ст.

BS: Уровень глюкозы в крови выражается в молярной концентрации, ммоль/л.

BodyTemp: температура тела в градусах по фаренгейту

HeartRate: Нормальная частота пульса в состоянии покоя в ударах в минуту.

Risk Level: Прогнозируемый уровень интенсивности риска во времущего признака.

```python
%pip install seaborn
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

from sklearn.model_selection import train_test_split, RandomizedSearchCV, cross_val_score
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import LabelEncoder, StandardScaler, MinMaxScaler
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.metrics import accuracy_score
from sklearn.metrics import multilabel_confusion_matrix
from sklearn.metrics import recall_score

from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from lightgbm import LGBMClassifier
```

```python
num_columns = ['Age', 'SystolicBP', 'DiastolicBP', 'BS', 'BodyTemp', 'HeartRate']
def boxplot_func(data, num_columns):
    fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 5))
    for i, column in enumerate(num_columns):
        row_index = i // 3
        col_index = i % 3
        data.boxplot(column=column, ax=axes[row_index, col_index])
boxplot_func(df, num_columns)
```

```python
RANDOM_STATE = 42
TEST_SIZE = 0.25

X = df.drop(['RiskLevel'], axis=1)
y = df['RiskLevel']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = TEST_SIZE, random_state = RANDOM_STATE)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
#закодируем целевой признак
label_encoder = LabelEncoder()
 
y_train_coded = label_encoder.fit_transform(y_train)
y_test_coded = label_encoder.transform(y_test)


# создаём общий пайплайн для подготовки данных
data_preprocessor = ColumnTransformer(
    [
        ('num', StandardScaler(), num_columns)
    ], 
    remainder='passthrough')

# создаём итоговый пайплайн: подготовка данных и модель
pipe_final= Pipeline(
    [
        ('preprocessor', data_preprocessor),
        ('models', DecisionTreeClassifier(random_state=RANDOM_STATE))
    ]
)

param_grid = [
    
    # словарь для модели DecisionTreeClassifier
    {'models': [DecisionTreeClassifier(random_state=RANDOM_STATE)],
     'models__max_depth': range(2,5),
     'models__max_features': range(2,5),
     'preprocessor__num': [StandardScaler(), 'passthrough']},
    
     # словарь для модели KNeighborsClassifier
    {'models': [KNeighborsClassifier()],
     'models__n_neighbors': range(2,5),
     'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']},
    
    # словарь для модели LogisticRegression
    {'models': [LogisticRegression(
        random_state=RANDOM_STATE, 
        solver='liblinear', 
        penalty='l1')],
     'models__C': range(1,5),
     'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']},

    # словарь для SCV
    {'models': [SVC(probability=True)],
     'models__C':[0.1,1,10],
     'models__kernel':['linear'],
     'models__degree':[1,3,4],
     'models__gamma':[1,0.1,0.01],
     'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']},

    # словарь для LGBMClassifier
    {'models': [LGBMClassifier(random_state=RANDOM_STATE)],
     'models__learning_rate':[0.1,0.3,0.5],
     'models__num_leaves': range(15, 50),
     'models__max_depth': range(2, 10),
     'preprocessor__num': [StandardScaler(), MinMaxScaler(), 'passthrough']}
]

rs = RandomizedSearchCV(
    pipe_final, 
    param_grid, 
    cv=5, 
    scoring='roc_auc_ovo',
    n_jobs=-1,
    error_score='raise'
)
rs.fit(X_train, y_train_coded)

print('Лучшая модель и её параметры:\n\n', rs.best_estimator_)
print ('Метрика лучшей модели на тренировочной выборке:', rs.best_score_)
```

```python
y_pred = rs.best_estimator_.predict(X_test) 
print(f'Точность модали на тестовой выборке: {round(accuracy_score(y_test_coded, y_pred), 2)}')
```

```python
# можно предположить, что сильнее всего нас интересует правильно присвоенные и определенные объекты с high risk
```

```python
multilabel_confusion_matrix(y_test_coded, y_pred, labels=[0, 1, 2])
```
